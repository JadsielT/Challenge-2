### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_tiendas = {'Tienda_1':0,'Tienda_2':0,'Tienda_3':0,'Tienda_4':0}
facturacion_tiendas['Tienda_1']=(sum(tienda['Precio']))
facturacion_tiendas['Tienda_2']=(sum(tienda2['Precio']))
facturacion_tiendas['Tienda_3']=(sum(tienda3['Precio']))
facturacion_tiendas['Tienda_4']=(sum(tienda4['Precio']))
print(facturacion_tiendas)
```

```python
claves = list(facturacion_tiendas.keys())
valores = list(facturacion_tiendas.values())
plt.bar(claves, valores)
plt.title("Facturacion por tienda")
plt.show()
```

# 2. Ventas por categoría

```python



df_group_1=tienda.groupby('Categoría del Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_2=tienda2.groupby('Categoría del Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_3=tienda3.groupby('Categoría del Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_4=tienda4.groupby('Categoría del Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_1=df_group_1.reset_index()
df_group_2=df_group_2.reset_index()
df_group_3=df_group_3.reset_index()
df_group_4=df_group_4.reset_index()


```

```python
Categoria = list(df_group_1['Categoría del Producto'])
Cantidad = list(df_group_1['Cantidad de productos'])
plt.pie(Cantidad, labels = Categoria, autopct="%0.1f %%")
plt.title("Tienda 1 categorias vendidas")
plt.show()

Categoria = list(df_group_2['Categoría del Producto'])
Cantidad = list(df_group_2['Cantidad de productos'])
plt.pie(Cantidad, labels = Categoria, autopct="%0.1f %%")
plt.title("Tienda 2 categorias vendidas")
plt.show()

Categoria = list(df_group_3['Categoría del Producto'])
Cantidad = list(df_group_3['Cantidad de productos'])
plt.pie(Cantidad, labels = Categoria, autopct="%0.1f %%")
plt.title("Tienda 3 categorias vendidad")
plt.show()

Categoria = list(df_group_4['Categoría del Producto'])
Cantidad = list(df_group_4['Cantidad de productos'])
plt.pie(Cantidad, labels = Categoria, autopct="%0.1f %%")
plt.title("Tienda 4 categorias vendidad")
plt.show()
```

# 3. Calificación promedio de la tienda


```python

Calificacion_tiendas = {'Calificacion_tienda_1':0,'Calificacion_tienda_2':0,'Calificacion_tienda_3':0,'Calificacion_tienda_4':0}
Calificacion_tiendas['Calificacion_tienda_1']=tienda['Calificación'].mean()
Calificacion_tiendas['Calificacion_tienda_2']=tienda2['Calificación'].mean()
Calificacion_tiendas['Calificacion_tienda_3']=tienda3['Calificación'].mean()
Calificacion_tiendas['Calificacion_tienda_4']=tienda4['Calificación'].mean()
print(Calificacion_tiendas)
```

```python
claves = list(Calificacion_tiendas.keys())
valores = list(Calificacion_tiendas.values())
plt.barh(claves, valores)
plt.xlim(right=5)
plt.title("Calificacion de las tiendas")
plt.show()
```

# 4. Productos más y menos vendidos

```python

df_group_1=tienda.groupby('Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_2=tienda2.groupby('Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_3=tienda3.groupby('Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_4=tienda4.groupby('Producto').size().to_frame('Cantidad de productos').sort_values(by='Cantidad de productos',ascending=False)

df_group_1=df_group_1.reset_index()
df_group_2=df_group_2.reset_index()
df_group_3=df_group_3.reset_index()
df_group_4=df_group_4.reset_index()
df_group_1.head()
```

```python

df_graf1=df_group_1.head()
df_graf1=pd.concat([df_graf1,df_group_1.tail()], ignore_index=True)
Producto = list(df_graf1['Producto'])
Cantidad = list(df_graf1['Cantidad de productos'])
plt.barh(Producto, Cantidad)
plt.title("productos mas vendidos y menos vendidos tienda 1")
plt.show()

df_graf2=df_group_2.head()
df_graf2=pd.concat([df_graf2,df_group_2.tail()], ignore_index=True)
Producto = list(df_graf2['Producto'])
Cantidad = list(df_graf2['Cantidad de productos'])
plt.barh(Producto, Cantidad)
plt.title("productos mas vendidos y menos vendidos tienda 2")
plt.show()

df_graf3=df_group_3.head()
df_graf3=pd.concat([df_graf3,df_group_3.tail()], ignore_index=True)
Producto = list(df_graf3['Producto'])
Cantidad = list(df_graf3['Cantidad de productos'])
plt.barh(Producto, Cantidad)
plt.title("productos mas vendidos y menos vendidos tienda 3")
plt.show()

df_graf4=df_group_4.head()
df_graf4=pd.concat([df_graf4,df_group_4.tail()], ignore_index=True)
Producto = list(df_graf4['Producto'])
Cantidad = list(df_graf4['Cantidad de productos'])
plt.barh(Producto, Cantidad)
plt.title("productos mas vendidos y menos vendidos tienda 4")
plt.show()
```

# 5. Envío promedio por tienda

```python

Envios_tiendas = {'Envios_tienda_1':0,'Envios_tienda_2':0,'Envios_tienda_3':0,'Envios_tienda_4':0}
Envios_tiendas['Envios_tienda_1']=tienda['Costo de envío'].mean()
Envios_tiendas['Envios_tienda_2']=tienda2['Costo de envío'].mean()
Envios_tiendas['Envios_tienda_3']=tienda3['Costo de envío'].mean()
Envios_tiendas['Envios_tienda_4']=tienda4['Costo de envío'].mean()
print(Envios_tiendas)
```

```python
claves = list(Envios_tiendas.keys())
valores = list(Envios_tiendas.values())
plt.bar(claves, valores)
plt.title("Envio por tienda")

plt.show()
```
